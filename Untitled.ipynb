### CRACKING THIS BUSINESS PROBLEM 

Renting a house or an apartment is never easy. Whether you are a college student or a working professional, renting a place 
always seems like a daunting task that is often impulsive or risky. Rent is influenced by several factors.

In this challenge, participants will predict the house-rents using data science methods, machine learning, and hyperparameter 
tuning.

Train: 134683 rows x 21 columns 

Test: 57722 rows x 20 columns

Data description:

- 'Property_ID', = Unique id of the property

- 'room',  No. of rooms available

- 'layout_type',  ['BHK' 'RK']

- 'property_type', = ['Apartment' 'Studio Apartment' 'Independent Floor' 'Independent House' 'Villa' 'Penthouse']

- 'locality',  = Overall Locality description

- 'price', = Rent in INR

- 'area', = in Square Feet

- 'furnish_type', ['Semi-Furnished' 'Furnished' 'Unfurnished']

- 'bathroom',: No. of Bathrooms

- 'city', : ['Ahmedabad' 'Bangalore' 'Chennai' 'Delhi' 'Hyderabad' 'Kolkata' 'Mumbai' 'Pune']

- 'parking_spaces',: 1 or 0  : 1 = Yes, 0 = No

- 'floor',: Floor from 0 to 20

- 'pet_friendly', : 1 or 0  : 1 = Yes, 0 = No

- 'power_backup', : 1 or 0  : 1 = Yes, 0 = No

- 'washing_machine',: 1 or 0  : 1 = Yes, 0 = No

- 'air_conditioner', : 1 or 0  : 1 = Yes, 0 = No

- 'geyser/solar',: 1 or 0  : 1 = Yes, 0 = No

- 'security_deposit', Amount of security Deposit

- 'neighborhood', : Nearest Metro/Bus/Mall in meters

- 'CCTV/security' : CCTV available 1 or 0  : 1 = Yes, 0 = No

- 'lift': 1 or 0  : 1 = Yes, 0 = No

```python
# distance < 1 : proximity , distance < 2 : vicinity 

def dist(x):
    if (x < 1):
        return('Proximity')
    elif(x < 2):
        return('Vicinity')
    else:
        return('Far')
        
```

```python
combined['distance'] = combined['neighbourhood'].apply(dist)
```

```python
combined.distance.value_counts()
```

```python
# room into category 1: studio, <3 small , bigger

def rooms(x):
    if(x == 1):
        return('Studio')
    elif(x < 3):
        return('Small')
    else:
        return('Large')
```

```python
combined['room_cat'] = combined['room'].apply(rooms)
```

```python
north = ['Delhi']
south  = ['Chennai', 'Bangalore', 'Hyderabad']
west = ['Mumbai', 'Ahmedabad', 'Pune']
east = ['Kolkata']
```

```python
def city_c(x):
    if x in north:
        return('North')
    elif(x in west):
        return('West')
    elif(x in east):
        return('East')
    else:
        return('South')
```

```python
combined['city_cat'] = combined['city'].apply(city_c)
```

```python
newtrain.shape, newtest.shape
```

```python
newtest = newtest.drop('price', axis = 1)
```

```python
dummytrain = pd.get_dummies(newtrain, drop_first = True)
dummytest = pd.get_dummies(newtest, drop_first = True)
```

# MODEL BUILDING

```python
#RANDOM FOREST MODEL

from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()

X = dummytrain.drop('price', axis= 1)
y = dummytrain.price
```

```python
price_rf = rf.fit(X,y).predict(dummytest)
```

```python
solution1 = pd.DataFrame({"price ": price_rf})
solution1.to_csv('RF_model.csv', index = False)
```

#### RMSE VALUE AFTER UPLOADING RANDOM FOREST CSV FILE IS 62448.04648

```python
from sklearn.ensemble import GradientBoostingRegressor

gbm = GradientBoostingRegressor()

X = dummytrain.drop('price', axis= 1)
y = dummytrain.price

price_gbm = gbm.fit(X,y).predict(dummytest)
```

```python
solution = pd.DataFrame({"price ": price_gbm})
solution.to_excel('GBM_model.xlsx', index = False)
```
